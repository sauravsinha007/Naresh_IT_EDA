**Data Transformation Technique**

   - Generally used for to convert Normal distribution

   - It is also avoid skew ness also

   - We have some important transformation technique:
    
        - Log Transformation
        
        - Reciprocal transformation
    
        - Square root transformation
    
        - Power transformation

$Step-1$

**Read the required packages**

```python
import numpy as np
import matplotlib.pyplot as plt
```

$Step-2$

**Read the data**

$Exponential-Data$

```python
exp_data = np.random.exponential(size=1000)
exp_data[:10]

# We are consider exponential data set with eandom 1k observations
# using numpy package
```

```python
plt.hist(exp_data,
        bins=30,
        label="Exponential")
plt.legend()
plt.show()
```

$Norm-Data$

```python
norm_data = np.random.normal(size=1000)
plt.hist(norm_data)
```

$Step-3$

**Log Transformation**

- np.log is used for log transformation

- Generally log transformation will not convert data into normal

- It avoids skew ness

- n.log means natural logrithm base = e




```python
log base 2 called bits

log base 10

log base e calles ln

convert log base e to log base 10
 2.3 * 
    
    logek = 10
    k = e^10
    
    x = 10
    log 10 k = 10 --> k = 10^10
    
    log 2 k = 10 ----> k = 2 ^ 10
    
    log e k = 10 ---> k = e ^ 10
    
    
```

```python
log_data = np.log(exp_data)
log_data[:10]
```

```python
exp_data[:10]
```

```python
plt.hist(log_data,
        bins=40,
        label='Log Data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,
        bins=30,
        label="Exponential")
plt.legend()
plt.subplot(1,2,2).hist(log_data,
        bins=40,
        label='Log Transformation')
plt.legend()
plt.show()
```

$Step-4$

**Reciprocal transformation**

- Reciprocal transformation is fail if data has zero value

```python
# Take the exponential data
# rec_data = 1/exp_data
# plot the histogram and check it

rec_data = np.reciprocal(exp_data)
rec_data[:10]
```

```python
plt.hist(rec_data,
        bins= 40,
        label='Reciprocal transformation')
plt.legend()
plt.show()
```

$Step-5$
**Square root transformation**

```python
print(25**2) # square
print(25 ** (1/2)) ## square root
print(np.sqrt(25))
```

```python
sqrt_data = np.sqrt(exp_data)
sqrt_data[:10]
```

```python
plt.hist(sqrt_data,
        bins=40,
        label="Square root data")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,
        bins=30,
        label="Exponential")
plt.legend()
plt.subplot(1,2,2).hist(sqrt_data,
        bins=40,
        label="Square root data")
plt.legend()
plt.show()
```

```python
import seaborn as sns
sns.displot(sqrt_data)
```

$Step-6$

**Power transformation**

- It is related to sklearn package

- Package name: sklearn.preprocessing

- Method name: Power Transformer

- Inside Box-Cox, yeo-jhonson

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer()
# by default it take yeo-jhonson
pt_data = pt.fit_transform([exp_data])
pt_data
```

```python
plt.hist(pt_data)
```

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer()

pt_data1 = pt.fit([exp_data]).transform([exp_data])
```

```python
pt_data1
```

```python
plt.hist(pt_data1[0])
plt.show()
```
