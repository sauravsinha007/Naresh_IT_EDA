**Method-1**

$map$

- First consider one categorical column

- Observe how many unique labels are there

- Assume here we are taking case_status column

- case_status 2 unique labels are there
    - Dined
    - Certified
    
-Create a dictionary by providing key:value

- Here key= Dined and the value = 1
- key = Certified and the value = 0

- dictionary will look like: {'Certified' : 0,
                             'Denied': 1}



```python
visa_df['case_status'].unique()
```

```python
d1 = {'Certified' : 0, 'Denied': 1}
visa_df['case_status'].map(d1)

# After mapping, in which column you want to apply the data
# you want to create a new column
# you want to use old column

# overriding is recommended
visa_df['case_status'] = visa_df['case_status'].map(d1)
```

```python
visa_df.head()
```

```python
visa_df['continent'].unique()
```

```python
d1 = {}
labels = visa_df['continent'].unique()
for i in range(len(labels)):
    d1[labels[i]] = i

d1
```

```python
visa_df['continent'] = visa_df['continent'].map(d1)
visa_df.head()
```

- np.where is kind of if-else

- np.where is useful for binary condition

- which mean it is useful for the columns having two labels

- np.where will take 3 parameters

    - condition
    
    - True value
    
    - False value
    
- I want to replace case status column with 1 when it is Denied

- Otherwise it is zero

- con = visa_df['case_status'] === 'Denied'

- True value -= 1   False value = 0 

- np.where(con , <True value>, <false Value>)

```python
con = visa_df['case_status'] == 'Denied'
visa_df['case_status'] = np.where(con, 1,0)
visa_df.head()
```

**Method-3**

$one-hot-encoder$

pd.getdummies()

- one hot encoder means at a time one will be ON, Other is OFF

- ON represents with 1 and Off represents with 0

- One hot encoder creates new extra columns based on labels in the columns

- For example case_status has 2 unique labels

- If you apply one hot encoder on case_status it creates 2 extra columns

    - case_status_Denied
    - case_status_Certified


|Case status|case_status_Denied|case_status_Certified
-------------------------------------------------
|Denied        |   1               |  0
|Certified     |   0               |  1

**Advantage****

- One hot encoder gives independency b/w the varoavles

- It is one of the important property before apply ML models

- Variables are independent each other

- Variables are perpendicular each other

- Variables are Orthogonl each other

- Variables are 90 degree pahse shift

- **Disadvantage**

- Assume that column has 100 unique labels

- If we apply hot encoder it will create 100 New Columns

- Your data matrix become sparse

- We required More time and more menory to compute the calculation

- This is called Curse of Dimension
